```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing students social media addictions dataset
df=pd.read_csv('C:/Users/HP/Downloads/Students Social Media Addiction.csv')
print("\nDataset loaded successfully\n")
```

```python
print("Dataset shape is:", df.shape)
```

```python
print("columns data types are:")
df.dtypes
```

```python
df.head() 
```

```python
#Finding null values
df.isna().sum()
```

```python
#Dropping null values from the dataset
df_cleaned=df.dropna()
df_cleaned
```

```python
#filtering numerical columns
numeric_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols
```

```python
#Basic statistics of the numerical columns
df_cleaned[numeric_cols].describe()
```

```python
# Group by Gender and getting the mean of numerical columns
grouped_by_gender = df_cleaned.groupby("Gender").mean(numeric_only=True)
print("Mean values grouped by Gender:")
print(grouped_by_gender)
```

Females:

Spend more time on social media (avg: 5.01 hrs/day) than males (avg: 4.82 hrs/day).

Report more conflicts over social media.

Have a slightly higher Addicted Score (6.51) than males (6.35).

Sleep slightly less on average than males

 Insight: Females may be slightly more affected by or engaged with social media, which could correlate with increased conflict and addictive tendencies.



```python
# Group by Academic Level and getting the mean Addicted Score
grouped_by_academic = df_cleaned.groupby("Academic_Level")["Addicted_Score"].mean()
print("\nMean Addicted Score grouped by Academic Level:")
print(grouped_by_academic)
```

High School students have the highest Addicted Score (mean: 8.04).

Undergraduates: moderately high (mean: 6.49)

Graduates: lowest (mean: 6.24)

 Insight: Younger students (high schoolers) appear to be more prone to social media addiction. This could be due to factors like emotional development, peer pressure, or fewer academic/work responsibilities.



```python
grouped_by_platform = df_cleaned.groupby("Most_Used_Platform").mean(numeric_only=True)
print("\nMean values grouped by Most Used Platform:")
print(grouped_by_platform)
```

WhatsApp, Tiktok and Snapchat are most addictive with mean over 5.0

 Potential Insight: Some platforms may be more "addictive" due to their content format or engagement loops.

```python
sns.set(style="whitegrid")
sns.figsize(10, 6)
```

```python
# Line Chart: Trend of Average Daily Usage 

plt.plot(df_cleaned.index, df_cleaned["Avg_Daily_Usage_Hours"], color="blue")
plt.title("Trend of Average Daily Social Media Usage ")
plt.xlabel("Entry ")
plt.ylabel("Avg Daily Usage (Hours)")
plt.tight_layout()
plt.show()
```

```python
# Bar Chart: Average Addicted Score by Academic Level
plt.figure(figsize=(8, 5))
sns.barplot(data=df_cleaned, x="Academic_Level", y="Addicted_Score", estimator='mean', ci=None, palette="pastel")
plt.title("Average Addicted Score by Academic Level")
plt.ylabel("Mean Addicted Score")
plt.xlabel("Academic Level")
plt.tight_layout()
plt.legend("Academic Level")
plt.show()
```

```python
# Histogram: Distribution of Mental Health Scores
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned["Mental_Health_Score"], bins=10, kde=True, color="blue")
plt.title("Distribution of Mental Health Scores")
plt.xlabel("Mental Health Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Daily Usage vs Addicted Score
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_cleaned, x="Avg_Daily_Usage_Hours", y="Addicted_Score", hue="Gender", alpha=0.7)
plt.title("Addicted Score vs Average Daily Social Media Usage")
plt.xlabel("Avg Daily Usage (Hours)")
plt.ylabel("Addicted Score")
plt.tight_layout()
plt.show()
```
